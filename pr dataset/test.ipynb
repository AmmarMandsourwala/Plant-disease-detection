```python
import os 
def predict(model, img_path, class_names):
    img=tf.keras.preprocessing.image.load_img(img_path, target_size=(128,128 ))  # resize as per model input
    img_array=tf.keras.preprocessing.image.img_to_array(img)
    img_array=tf.expand_dims(img_array, 0)  # Adding batch dimension

    predictions=model.predict(img_array)
    predicted_class=class_names[np.argmax(predictions[0])]

    return predicted_class
```

```python
image_dir = 'test/test'

for filename in os.listdir(image_dir):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # check for image extensions
        img_path = os.path.join(image_dir, filename)
        predicted_class= predict(model, img_path, class_name)

        # Displaying image 
        img=cv2.imread(img_path)
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(2,2))
        plt.imshow(img)
        plt.title(f"{filename} -> Predicted: {predicted_class}")
        plt.xticks([])
        plt.yticks([])
        plt.show()
        
        
```
